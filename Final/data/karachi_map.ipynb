```python
from math import radians, cos, sin, asin, sqrt

# Define Haversine distance function
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371  # Earth radius in kilometers
    return c * r

# Approximate agricultural locations around Karachi
agri_areas = [
    {'name': 'Malir District', 'lat': 24.9000, 'lon': 67.1500},
    {'name': 'Gadap Town', 'lat': 25.0500, 'lon': 67.2000},
    {'name': 'Dhabeji', 'lat': 24.7833, 'lon': 67.5000},
    {'name': 'Thatta District', 'lat': 24.7500, 'lon': 67.9000}
]

# Function to check if a sensor is near any agri area
def is_near_agriculture(sensor_lat, sensor_lon, threshold_km=10):
    for agri in agri_areas:
        dist = haversine(sensor_lon, sensor_lat, agri['lon'], agri['lat'])
        if dist <= threshold_km:
            return True
    return False

# Load your CSV (adjust path as needed)
df = PAQI
df.columns = df.columns.str.strip()  # Remove any accidental whitespace

# Tag each sensor row
df['near_agriculture'] = df.apply(
    lambda row: is_near_agriculture(row['latitude'], row['longitude']),
    axis=1
)

# Optional: Save or inspect the tagged data
df.to_csv("tagged_sensors.csv", index=False)
print(df[['Name', 'latitude', 'longitude', 'near_agriculture']].drop_duplicates())
```
